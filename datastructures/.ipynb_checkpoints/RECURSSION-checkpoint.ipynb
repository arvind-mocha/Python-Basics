# TYPES OF RECURSSION

```python
# HEAD RECURSSION
# RECURSSION ON THE BEGINING OF THE STATEMENT

def cal(val):
    if val > 0:
        cal(val-1)
        k = val**2
        print(k)
        
cal(4)
```

```python
# TAIL RECURSSION
# RECURSSION ON THE END OF THE STATEMENT

def cal(val):
    if val > 0:
        k = val**2
        print(k)
        cal(val-1)
        
cal(4)
```

EXPLANATION ( DRY RUN ):

val = 4; val > 0 TRUE; 

val = 1; k = 1 ** 2 = 1;cal(val-1)

val = 3; val > 0 TRUE; 

val = 2; k = 1 ** 2 = 4;cal(val-1)

val = 2; val > 0 TRUE; cal(val-1)

val = 1; val > 0 TRUE; cal(val-1)

val = 0; val > 0 FALSE;

NOW THE FUNCTION STARTS ITS EXECUTION FROM WHERE IT STOPPED THE RECURSSION





val = 3; k = 1 ** 2 = 9

val = 4; k = 1 ** 2 = 16

# FACTORIAL RECURSSION

```python
%%timeit
def fact(val):
    
    if val == 1 :
        return val
    else:
        return val*fact(val-1)
    
print(fact(5))
```

```python
%%timeit
def fact(val):
    ans = 1
    for i in range(val):
        ans = val * ans
        val -= 1
    return ans
print(fact(5))
```
