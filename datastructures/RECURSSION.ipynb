# TYPES OF RECURSSION

```python
# HEAD RECURSSION
# RECURSSION ON THE BEGINING OF THE STATEMENT

def cal(val):
    if val > 0:
        cal(val-1)
        k = val**2
        print(k)
        
cal(4)
```

```python
# TAIL RECURSSION
# RECURSSION ON THE END OF THE STATEMENT

def cal(val):
    if val > 0:
        k = val**2
        print(k)
        cal(val-1)
        
cal(4)
```

EXPLANATION ( DRY RUN ):

val = 4; val > 0 TRUE; 
k = 4 ** 2 = 16; cal(val-1)

val = 3; val > 0 TRUE; 
k = 3 ** 2 = 9 ; cal(val-1)

val = 2; val > 0 TRUE; 
k = 2 ** 2 = 4; cal(val-1)

val = 1; val > 0 TRUE;
val = 1; k = 1 ** 2 = 1

val = 0; val > 0 FALSE;

# TREE RECURSSION

```python
# CALLING THE FUNCTION INSIDE A FUNCTION MORE THAN ONCE
```

```python
def cal(val):
    if val > 0:
        cal(val-1)
        k = val**2
        print(k)
        cal(val-1)
        
cal(3)
```

# FACTORIAL RECURSSION

```python
def fact(val):
    
    if val == 1 :
        return val
    else:
        return val*fact(val-1)
    
print(fact(5))
```

```python
def fact(val):
    ans = 1
    for i in range(val):
        ans = val * ans
        val -= 1
    return ans
print(fact(5))
```
